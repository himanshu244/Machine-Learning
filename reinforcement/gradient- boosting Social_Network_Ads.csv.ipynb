```python
import warnings
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
warnings.filterwarnings("ignore")
df=pd.read_csv("Social_Network_Ads.csv")
x=df.iloc[:,[1,2,3]]
y=df.iloc[:,4:]
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
le=LabelEncoder()#approach 1,2,3
x['Gender']=le.fit_transform(x['Gender'])
from sklearn.model_selection import train_test_split as tts
x_train,x_test,y_train,y_test=tts(x,y,test_size=0.2,random_state=2)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
gb=GradientBoostingClassifier(n_estimators=50,max_depth=3,learning_rate=0.1)
gb.fit(x_train,y_train)
```

```python
h=gb.predict(x_test)
from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(y_test,h)
print(cm)
k=accuracy_score(y_test,h)
k
```

```python
from sklearn.model_selection import GridSearchCV
rf1=GradientBoostingClassifier()
d={'n_estimators':[10,40,100],'max_depth':[10,20,30,None],'learning_rate':[0.01,0.1,1]}
gs=GridSearchCV(rf1,d,cv=5)#cv-crossvalidation(k fold)
gs.fit(x,y)
```

```python
dff=pd.DataFrame(gs.cv_results_)
dff
```
