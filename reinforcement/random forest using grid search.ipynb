```python
from sklearn.metrics import precision_recall_fscore_support as prfs
from sklearn.metrics import confusion_matrix,accuracy_score
def grid(depth,estimator):
    rf=RandomForestClassifier(criterion='entropy',n_estimators=estimator,max_depth=depth)
    rf.fit(x_train,y_train)
    h=rf.predict(x_test)
    precision,recall_,fscore,support=prfs(y_test.values,h,pos_label=1,average='binary')
    accuracy=accuracy_score(y_test,h)
    print("Est=",estimator,"Depth=",depth,"Precision=",precision,"Acccuracy=",accuracy,"Recall=",recall_)
```

```python
estimators=[10,50,100]
depth=[10,20,30,None]
for e in estimators:
    for d in depth:
        grid(d,e)
```
