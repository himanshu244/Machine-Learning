```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
df=pd.read_csv("Market_Basket_Optimisation.csv",header=None)
from apyori import apriori
```

```python
df.head()
```

```python
df.info()
```

```python
transactions=[]
for i in range(7501):
    """ tmp_lst=[]
    for j in range(20):
        tmp_lst.append(str(df.values[i,j]))
    transactions.append(tmp_lst)
    tmp_lst = []"""
    transactions.append([str(df.values[i,j]) for j in range(20)])
```

```python
rules=apriori(transactions,min_support=0.003,min_confidence=0.2,min_lift=3,max_length=10)
```

```python
l=list(rules)
print(l)
```

```python
print(l[0])#gives 1st row
print(l[0][0])#gives frozen set of item
item=list(l[0][0])
print(item[0],item[1])
print(l[0][1])#support
print(l[0][2][0])
print(l[0][2][0][2])#confidence
print(l[0][2][0][3])#lift
len(l)
```

```python
support=[]
items=[]
confidence=[]
lift=[]
for i in range(154):
    support.append(l[i][1])
    confidence.append(l[i][2][0][2])
    lift.append(l[i][2][0][3])
    items.append(list(l[i][0]))
```

```python
items
```

```python
df=pd.DataFrame({'items':items,'support':support,'confidence':confidence,'lift':lift})
```
