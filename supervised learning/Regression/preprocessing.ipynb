# categorical data

```python
pd.get_dummies(df['day']).head()
```

```python
d=df['day'].unique()
for i in d:
    j=[]

    for row in df['day']:
        if row == i :    
            j.append(1) 
        else:
            j.append(0)
    df[i]=j
```

```python
df.sample(10)
```

# encoding

```python
df['day']=df['day'].replace({'Thur':1,'Fri':2,"Sat":2,"Sun":3})
```

```python
df.sample(5)
```

# standard scaling

```python
m1=df['total_bill'].max()
m2=df['tip'].max()
df['total_bill']=df['total_bill']/m1
df['tip']=df['tip']/m2
df.head()
```

```python
m1=df['total_bill'].min()
m2=df['tip'].min()
print(m1,m2)
```
